```python
import pandas as pd
import matplotlib.pyplot as plt  # Make sure this line is present
from sklearn.preprocessing import MinMaxScaler

file_path = r"C:\Users\eloik\.jupyter\nyc_taxi_fare_cleaned.csv"
df = pd.read_csv(file_path)
```

```python
stats = df.describe().T
print(stats)

plt.figure(figsize=(10, 6))
plt.hist(df["fare_amount"], bins=100, edgecolor="black", alpha=0.7, log=True)
plt.xlabel("Fare Amount (USD)")
plt.ylabel("Frequency (log scale)")
plt.title("Histogram of Taxi Fare Amounts")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
